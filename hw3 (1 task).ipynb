```python
import seaborn as sns
import pandas as pd
from sklearn.datasets import make_circles
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler
%matplotlib inline
```

## 1-ое задание

#### 1-ый набор данных

```python
x,y = make_circles(n_samples=100, noise=0.25,factor=0.5)
plt.scatter(x[:,0], x[:,1], c=y, edgecolors= 'b')
plt.clf
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, random_state=5)
scaler = StandardScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
x_train.shape
```

```python
train_x = pd.DataFrame(x_train, index=x_train[:,0])
train_x
```

```python
#KNN (k=7), train
knn_x = KNeighborsClassifier(n_neighbors=7)
knn_x.fit(x_train, y_train)
```

```python
knn_x.score(x_train, y_train)
```

```python
knn_x_predictions = knn_x.predict(x_train)
knn_x_predictions
```

```python
confusion_matrix(y_train, knn_x_predictions)
```

```python
print(classification_report(y_train, knn_x_predictions))
```

```python
#KNN (k=4), test
knn_x = KNeighborsClassifier(n_neighbors=4)
knn_x.fit(x_test, y_test)
```

```python
knn_x.score(x_test, y_test)
```

```python
knn_x_predictions = knn_x.predict(x_test)
knn_x_predictions
```

```python
confusion_matrix(y_test, knn_x_predictions)
```

```python
print(classification_report(y_test, knn_x_predictions))
```
